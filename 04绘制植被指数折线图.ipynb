# 绘制指数指标图

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# dfa = pd.read_csv(r'E:\Desktop\Summary\种植结构提取-水稻\datas\Sample\ndvi_ndgi_rvi.csv', header=0, index_col=0)
# dfa = pd.read_csv(r'E:\Desktop\Summary\种植结构提取-水稻\datas\Sample1\ndvi_ndgi_rvi.csv', header=0, index_col=0)
dfa = pd.read_csv(r'E:\Desktop\Summary\种植结构提取-水稻\datas\Sample_corn\corn_ndvi_ndgi_rvi.csv', header=0, index_col=0)
# 创建数据
df = pd.DataFrame({'x': [0, 1, 2, 3, 4, 5],
                   'NDVI': dfa.iloc[0].tolist(),
                   'NDGI': dfa.iloc[1].tolist(),
                   'RVI': dfa.iloc[2].tolist()})
print(df)

# 绘制折线图
# df.plot(x='x', y=['NDVI', 'NDGI', 'RVI'])

# 获取折线图的数据
x_data = df['x'].values
print(x_data)
NDVI_data = df['NDVI'].values
NDGI_data = df['NDGI'].values
RVI_data = df['RVI'].values

# 计算斜率
slope_ndvi = np.gradient(NDVI_data, x_data)
slope_ndgi = np.gradient(NDGI_data, x_data)
slope_rvi = np.gradient(RVI_data, x_data)

# 打印斜率
print('slope_ndvi:\n', slope_ndvi, '\nslope_ndgi:\n', slope_ndgi, '\nslope_rvi:\n', slope_rvi)
```

```python
vi = pd.DataFrame(slope_ndvi, columns={"slope_ndvi"})
vi['slope_ndgi'] = slope_ndgi
vi['slope_rvi'] = slope_rvi
vi['x'] = ['07-03', '08-02', '08-22', '09-16', '10-06', '10-21']
vi
```
